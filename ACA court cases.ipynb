```python
import pandas as pd

# Specify the file path
file_path = r"C:\Users\z5388659\OneDrive - UNSW\Desktop\jupyter\Data\20230327_Environment-ACA court cases.xlsx"

# Read all sheets in the Excel file
all_sheets = pd.read_excel(file_path, engine='openpyxl', sheet_name=None)

# Combine all sheets into a single DataFrame using pandas.concat
combined_df = pd.concat(all_sheets.values(), ignore_index=True)

# Display the combined data
print(combined_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming combined_df is already loaded

# Convert the "Date" column to datetime format
combined_df['ReleaseDate'] = pd.to_datetime(combined_df['ReleaseDate'])

# Set the "Date" column as the index
combined_df.set_index('ReleaseDate', inplace=True)

# Plot the different columns
columns_to_plot = ['Instrument', 'NumberOfPages', 'PriceSensitivity', 'ReportTypes']  # Replace these with your column names

for column in columns_to_plot:
    plt.plot(combined_df.index, combined_df[column], label=column)

plt.xlabel('ReleaseDate')
plt.ylabel('Values')
plt.title('Time-series Visualization')
plt.legend()
plt.show()
```
